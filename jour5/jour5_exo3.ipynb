## Exercice 3 - Jour 5 : Petit syst√®me de gestion de stock

```python
# jour5/exercice3.py
# Jour 5 - Dictionnaires et boucles

print("=== Gestion de stock ===")

# Dictionnaire repr√©sentant le stock d‚Äôun magasin
stock = {
    "pomme": 10,
    "banane": 8,
    "orange": 5
}
```

```python
# Boucle pour afficher le stock actuel
print("\nStock actuel :")
for produit, quantite in stock.items():
    print(f"{produit} : {quantite} en stock")
```

```python
# Ajouter un produit
produit_ajoute = input("\nQuel produit veux-tu ajouter ? ")
quantite_ajoutee = int(input("Combien ? "))
stock[produit_ajoute] = quantite_ajoutee
print(f"‚úÖ {produit_ajoute} ajout√© avec {quantite_ajoutee} unit√©s.")
```

```python
# Mettre √† jour la quantit√© d‚Äôun produit existant
produit_modif = input("\nQuel produit veux-tu mettre √† jour ? ")
if produit_modif in stock:
    nouvelle_qte = int(input("Nouvelle quantit√© : "))
    stock[produit_modif] = nouvelle_qte
    print(f"üîÅ Quantit√© de {produit_modif} mise √† jour √† {nouvelle_qte}.")
else:
    print("‚ùå Produit non trouv√©.")
```

```python
# Afficher le stock final
print("\n=== Stock final ===")
for produit, quantite in stock.items():
    print(f"{produit} : {quantite}")
```
