## Exercice 2 - Jour 5 : Parcourir un dictionnaire

```python
# jour5/exercice2.py
# Jour 5 - Boucles sur dictionnaire

# Dictionnaire contenant des notes d'étudiants
notes = {
    "Sefdine": 19,
    "Fatima": 17,
    "Youssef": 9,
    "Nadia": 12
}
```

```python
# Afficher chaque étudiant et sa note
print("=== Notes des étudiants ===")
for nom, note in notes.items():
    print(f"{nom} a eu {note}/20")
```

```python
# Calculer la moyenne de la classe
moyenne = sum(notes.values()) / len(notes)
print(f"\nMoyenne de la classe : {moyenne:.2f}")
```

```python
# Trouver les étudiants au-dessus de la moyenne
print("\nÉtudiants au-dessus de la moyenne :")
for nom, note in notes.items():
    if note > moyenne:
        print(f"- {nom} ({note}/20)")
```

> Ce que tu apprends :

* Utiliser `.items()` pour parcourir clé et valeur
* Extraire toutes les valeurs avec `.values()`
* Faire des calculs sur les valeurs
* Combiner conditions et boucles sur dictionnaire
