## Exercice 1 - Jour 5 : Créer et manipuler un dictionnaire

```python
# jour5/exercice1.py
# Jour 5 - Dictionnaires : bases

# Création d'un dictionnaire représentant une personne
personne = {
    "nom": "Sefdine",
    "age": "30",
    "ville": "Rehemani"
}
```

```python
# Affichage du dictionnaire complet
print("Informations de la personne :", personne)
```

```python
# Accéder à une valeur spécifique
print("Nom :", personne["nom"])
print("Âge :", personne["age"])
```

```python
# Ajouter une nouvelle clé/valeur
personne["profession"] = "Ingenieur statisticien"

# Modifier une valeur existante
personne["age"] = 29

# Supprimer une clé
del personne["ville"]

# Afficher le dictionnaire final
print("\nDictionnaire mis à jour :", personne)
```
